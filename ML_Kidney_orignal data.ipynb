````python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

real_df = remove_outliers(real_df, "bp")
````

# **Model Initialization**

````python
X = real_df.drop(columns='class')
y = real_df["class"]
````

````python
X.head()
````

````python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, accuracy_score
from sklearn.model_selection import StratifiedKFold, LeaveOneOut,ShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
````

````python
# Define a dictionary to store the models
models = {
    'Logistic Regression': ,
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(),
    'Neural Network': MLPClassifier()
}
# Initialize Leave-One-Out cross-validator
cv_methods = {
    'Stratified K-Fold': StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    'Leave-One-Out': LeaveOneOut(),
    'Shuffle-Split': ShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
}
````

````python
# Perform cross-validation for each model and each cross-validation method
results = {}
for model_name, model in models.items():
    print(f"Model: {model_name}")
    for cv_name, cv_method in cv_methods.items():
        # Lists to store evaluation metrics
        precision_scores = []
        recall_scores = []
        f1_scores = []
        accuracy_scores = []

        # Perform cross-validation
        for train_idx, test_idx in cv_method.split(X, y):
            X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
            y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

            # Train the model
            model.fit(X_train, y_train)
            # Make predictions
            y_pred = model.predict(X_test)

            # Calculate evaluation metrics
            precision = precision_score(y_test, y_pred)
            recall = recall_score(y_test, y_pred)
            f1 = f1_score(y_test, y_pred)
            accuracy = accuracy_score(y_test, y_pred)

            # Append scores to lists
            precision_scores.append(precision)
            recall_scores.append(recall)
            f1_scores.append(f1)
            accuracy_scores.append(accuracy)

        # Store results
        results[(model_name, cv_name)] = {
            'Precision': np.mean(precision_scores),
            'Recall': np.mean(recall_scores),
            'F1-score': np.mean(f1_scores),
            'Accuracy': np.mean(accuracy_scores)
        }

# Create DataFrame for results
results_df = pd.DataFrame(results).T
````

````python
# Create subplots for each evaluation metric
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
axes = axes.flatten()
metrics = ['Precision', 'Recall', 'F1-score', 'Accuracy']

for i, metric in enumerate(metrics):
    ax = axes[i]
    ax.set_title(f'{metric} Comparison')
    ax.set_xlabel('Cross-Validation Method')
    ax.set_ylabel('Average Score')

    # Plot average scores for each model
    for model_name in models.keys():
        scores = results_df.loc[model_name, metric]
        ax.bar(results_df.index.levels[1], scores, label=model_name)

    ax.legend()

plt.tight_layout()
plt.show()
````

````python
from sklearn.metrics import roc_auc_score

# Initialize dictionary to store AUC scores
auc_results = {}

# Perform cross-validation for each model and each cross-validation method
for model_name, model in models.items():
    print(f"Model: {model_name}")
    for cv_name, cv_method in cv_methods.items():
        # Initialize list to store AUC scores for this model and cross-validation method
        auc_scores = []

        # Perform cross-validation
        for train_idx, test_idx in cv_method.split(X, y):
            X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
            y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

            # Check if both classes are present in the test set
            if len(np.unique(y_test)) == 2:
                # Train the model
                model.fit(X_train, y_train)
                # Make predictions
                y_pred_proba = model.predict_proba(X_test)[:, 1]  # Probability of positive class

                # Calculate AUC score
                auc = roc_auc_score(y_test, y_pred_proba)
                auc_scores.append(auc)
            else:
                print(f"Skipping AUC calculation for {model_name} - {cv_name} due to only one class present in y_true.")

        # Store average AUC score for this model and cross-validation method
        auc_results[(model_name, cv_name)] = np.mean(auc_scores)

        # Print average AUC score
        print(f"CV Method: {cv_name}")
        print(f"Average AUC: {np.mean(auc_scores)}")

# Convert results to DataFrame
auc_results_df = pd.DataFrame.from_dict(auc_results, orient='index', columns=['AUC'])

# Display AUC scores
print(auc_results_df)
````
